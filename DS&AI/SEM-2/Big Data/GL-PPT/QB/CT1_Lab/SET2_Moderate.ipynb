## Q1 (10 Marks) – Create DataFrame from CSV (Option A)
Create `students.csv` then read it using Spark DataFrame.

```python
# Create students.csv dataset
data = """name,marks
Alice,85
Bob,55
Charlie,72
Diana,45
Eve,91
Frank,63
Grace,38
Henry,77
Iris,58
Jack,88
"""

with open("students.csv", "w") as f:
    f.write(data)

print("students.csv created successfully")
print(data)
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("csv").getOrCreate()
spark.sparkContext.setLogLevel("ERROR")

df = spark.read.csv("students.csv", header=True, inferSchema=True)

print("=== DataFrame Contents ===")
df.show()

print("=== Schema ===")
df.printSchema()
```

---
## Q2 (5 Marks) – DataFrame Operations (Option A)
Filter students with marks > 60 and display name and marks.

```python
print("=== Students with marks > 60 ===")
df.filter(df.marks > 60).select("name", "marks").show()
```

## Q3 (5 Marks) – RDD Transformations (Option A)
Read `data.txt`, apply `flatMap → map → reduceByKey` to count word frequencies.

```python
# Create data.txt dataset
text = """hadoop spark hadoop
mapreduce spark hadoop
spark rdd dataframe
hadoop mapreduce spark
rdd spark hadoop
"""

with open("data.txt", "w") as f:
    f.write(text)

print("data.txt created successfully")
print(text)
```

```python
rdd = spark.sparkContext.textFile("data.txt")
result = rdd.flatMap(lambda x: x.split()).map(lambda x: (x, 1)).reduceByKey(lambda a, b: a + b)

print("=== Word Count Result ===")
print(result.collect())
```
