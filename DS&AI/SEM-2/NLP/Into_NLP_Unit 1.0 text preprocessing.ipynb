```python
# Import libraries
import nltk
import spacy
import re
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, LancasterStemmer
from nltk.stem import WordNetLemmatizer

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

print("✓ All libraries loaded successfully!")
```

## 1. Tokenization

**Tokenization** is the process of breaking down text into smaller units (tokens) like words or sentences.

### Types of Tokenization:
- **Word Tokenization**: Splits text into individual words
- **Sentence Tokenization**: Splits text into sentences
- **N-gram Tokenization**: Creates sequences of n words

```python
# Sample text
text = "Hello! This is a simple text preprocessing example. Let's see how tokenization works. It's really useful for NLP!"

print("Original Text:")
print(text)
print("\n" + "="*60 + "\n")

# Word Tokenization
word_tokens = word_tokenize(text)
print("Word Tokens:")
print(word_tokens)
print(f"\nTotal words: {len(word_tokens)}")

print("\n" + "="*60 + "\n")

# Sentence Tokenization
sent_tokens = sent_tokenize(text)
print("Sentence Tokens:")
for i, sent in enumerate(sent_tokens, 1):
    print(f"{i}. {sent}")
```

---

## 2. Stop Words Removal

**Stop Words** are common words that carry little semantic meaning (e.g., "the", "is", "a", "an").

⚠️ **Important Note**: Not always recommended to remove stop words! In some cases like sentiment analysis, words like "not", "no", "never" are crucial for understanding meaning.

```python
# Get English stop words
stop_words = set(stopwords.words('english'))

print(f"Total stop words in NLTK: {len(stop_words)}")
print("\nFirst 20 stop words:")
print(list(stop_words)[:20])
```

```python
# Example: Remove stop words
text = "I love Natural Language Processing"
tokens = word_tokenize(text.lower())

print("Original tokens:", tokens)

# Remove stop words
filtered_tokens = [word for word in tokens if word not in stop_words]

print("After removing stop words:", filtered_tokens)
print(f"\n✓ Note: 'I' is a stop word and was removed")
print(f"✓ Content words (love, natural, language, processing) are kept!")
```

---

## 3. Stemming

**Stemming** uses crude heuristic rules to chop off word endings.

### Characteristics:
- ⚡ Fast (rule-based)
- ❌ May produce non-words (e.g., "studies" → "studi")
- 🔄 Context-independent
- 📝 Good for: Search engines, approximate matching

### Popular Stemmers:
- **Porter Stemmer**: Most commonly used, less aggressive
- **Lancaster Stemmer**: More aggressive, can produce shorter stems

```python
# Initialize stemmers
porter = PorterStemmer()
lancaster = LancasterStemmer()

# Test words
words = ["running", "runs", "ran", "runner", "easily", "fairly", "studies", "studying", "caring", "better"]

print("Word".ljust(15), "Porter".ljust(15), "Lancaster")
print("="*50)

for word in words:
    porter_stem = porter.stem(word)
    lancaster_stem = lancaster.stem(word)
    print(f"{word.ljust(15)} {porter_stem.ljust(15)} {lancaster_stem}")
```

```python
# Stemming example with sentence
def stem_text(text):
    porter = PorterStemmer()
    
    # Tokenize and stem
    tokens = word_tokenize(text.lower())
    stemmed = [porter.stem(word) for word in tokens]
    
    print("Original:", tokens)
    print("Stemmed: ", stemmed)
    
    return ' '.join(stemmed)

text = "The runners are running in the park. They ran better than yesterday."
result = stem_text(text)
print("\nResult:", result)
```

## 4. Lemmatization

**Lemmatization** uses vocabulary and morphological analysis to return the dictionary base form.

### Characteristics:
- 🎯 Accurate (linguistic approach)
- ✅ Always produces valid dictionary words
- 🐌 Slower (requires dictionary lookup)
- 🧠 Context-aware (considers part of speech)
- 📝 Good for: Text analytics, NLP models requiring valid words

### NLTK vs spaCy Lemmatization:
- **NLTK**: Uses WordNet lemmatizer
- **spaCy**: More context-aware, considers POS tags automatically

```python
# NLTK Lemmatization
lemmatizer = WordNetLemmatizer()

words = ["running", "runs", "ran", "runner", "easily", "fairly", "studies", "studying", "caring", "better", "am", "is", "are"]

print("Word".ljust(15), "Lemma (noun)".ljust(15), "Lemma (verb)")
print("="*50)

for word in words:
    lemma_noun = lemmatizer.lemmatize(word, pos='n')  # noun
    lemma_verb = lemmatizer.lemmatize(word, pos='v')  # verb
    print(f"{word.ljust(15)} {lemma_noun.ljust(15)} {lemma_verb}")

print("\n⚠️ Note: NLTK lemmatizer requires POS tag to work properly!")
print("Without POS tag, 'better' stays 'better' instead of becoming 'good'")
```

```python
# spaCy Lemmatization (More powerful - automatically detects POS)
def lemmatize_with_spacy(text):
    doc = nlp(text)
    
    print("Token".ljust(15), "Lemma".ljust(15), "POS")
    print("="*50)
    
    for token in doc:
        print(f"{token.text.ljust(15)} {token.lemma_.ljust(15)} {token.pos_}")
    
    lemmatized = " ".join([token.lemma_ for token in doc])
    return lemmatized

text = "My name is Damir and I am running in the park. I have better running shoes than you."
print(f"Original: {text}\n")

result = lemmatize_with_spacy(text)
print(f"\nLemmatized Result: {result}")
```

## 5. Stemming vs Lemmatization Comparison

Let's compare both techniques side by side:

```python
def compare_stem_vs_lemma(text):
    # Stemming
    porter = PorterStemmer()
    tokens = word_tokenize(text.lower())
    stemmed = [porter.stem(word) for word in tokens]
    
    # Lemmatization
    doc = nlp(text)
    lemmatized = [token.lemma_ for token in doc]
    
    print("Original Text:", text)
    print("\n" + "="*60)
    print("\nOriginal Tokens:", tokens)
    print("\nStemmed (Porter):", stemmed)
    print("\nLemmatized (spaCy):", lemmatized)
    
    # Find differences
    print("\n" + "="*60)
    print("\nDifferences:")
    for orig, stem, lemma in zip(tokens, stemmed, lemmatized):
        if stem != lemma:
            print(f"  '{orig}' → Stem: '{stem}' | Lemma: '{lemma}'")

compare_stem_vs_lemma("The studies show that running improves health. Better results were seen in caring individuals.")
```

## 6. Complete Text Preprocessing Pipeline

Combining all techniques into a comprehensive preprocessing function:

```python
def preprocess_text(text, remove_stopwords=True, use_lemmatization=True):
    """
    Complete text preprocessing pipeline
    
    Steps:
    1. Lowercase
    2. Remove punctuation and special characters
    3. Tokenize
    4. Remove stop words (optional)
    5. Stemming or Lemmatization
    """
    print("="*60)
    print("TEXT PREPROCESSING PIPELINE")
    print("="*60)
    
    # Step 1: Lowercase
    text = text.lower()
    print("\n1. After Lowercasing:")
    print(f"   {text}")
    
    # Step 2: Remove punctuation
    text_clean = re.sub(r'[^\w\s]', '', text)
    print("\n2. After Removing Punctuation:")
    print(f"   {text_clean}")
    
    # Step 3: Tokenize
    tokens = word_tokenize(text_clean)
    print("\n3. After Tokenization:")
    print(f"   {tokens}")
    
    # Step 4: Remove stop words
    if remove_stopwords:
        stop_words = set(stopwords.words('english'))
        tokens = [word for word in tokens if word not in stop_words]
        print("\n4. After Stop Words Removal:")
        print(f"   {tokens}")
    
    # Step 5: Lemmatization or Stemming
    if use_lemmatization:
        doc = nlp(' '.join(tokens))
        tokens = [token.lemma_ for token in doc]
        print("\n5. After Lemmatization:")
        print(f"   {tokens}")
    else:
        porter = PorterStemmer()
        tokens = [porter.stem(word) for word in tokens]
        print("\n5. After Stemming:")
        print(f"   {tokens}")
    
    # Final result
    result = ' '.join(tokens)
    print("\n" + "="*60)
    print("FINAL RESULT:")
    print(f"   {result}")
    print("="*60)
    
    return result

# Test the pipeline
text = "Hello! This is a simple text preprocessing example. Let's see how it works with multiple sentences."
result = preprocess_text(text, remove_stopwords=True, use_lemmatization=True)
```

## 7. Additional Preprocessing Techniques

```python
# Other common preprocessing techniques

def advanced_preprocessing(text):
    print("Original Text:")
    print(text)
    print("\n" + "="*60 + "\n")
    
    # Remove URLs
    text = re.sub(r'http\S+|www.\S+', '', text)
    print("After removing URLs:")
    print(text)
    print()
    
    # Remove numbers
    text_no_nums = re.sub(r'\d+', '', text)
    print("After removing numbers:")
    print(text_no_nums)
    print()
    
    # Remove extra whitespace
    text_clean = re.sub(r'\s+', ' ', text_no_nums).strip()
    print("After removing extra spaces:")
    print(text_clean)
    print()
    
    # Remove special characters (keep only alphanumeric and spaces)
    text_final = re.sub(r'[^a-zA-Z\s]', '', text_clean)
    print("After removing special characters:")
    print(text_final)
    
    return text_final

sample = "Check this out: https://www.example.com! I scored 95% on my test. Amazing!!! 🎉"
result = advanced_preprocessing(sample)
```

## Summary

### Key Takeaways:

1. **Tokenization**: Breaking text into words/sentences - essential first step
2. **Stop Words**: Common words with little meaning - remove with caution!
3. **Stemming**: Fast but crude, may create non-words
4. **Lemmatization**: Accurate but slower, always creates valid words
5. **Choose wisely**: 
   - Use **Stemming** for: Speed-critical apps, search engines
   - Use **Lemmatization** for: Accuracy-critical apps, text analytics

### When to Apply Each Technique:

| Technique | Use Case |
|-----------|----------|
| Tokenization | Always needed as first step |
| Stop Words Removal | Text classification, topic modeling (NOT for sentiment analysis) |
| Stemming | Search engines, information retrieval |
| Lemmatization | Text analytics, machine translation, QA systems |
| Lowercasing | Almost always (except for NER, POS tagging) |
| Punctuation Removal | Depends on task (keep for sentiment) |

### Remember:
> **"There's no one-size-fits-all preprocessing pipeline. Choose techniques based on your specific NLP task and data characteristics!"**
