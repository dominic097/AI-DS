```python
import re
def remove_special_characters (text, remove_digits=False):
  #Using regex
  if remove_digits:
    pattern = r'[^a-zA-Z\s]'
  else:
    pattern = r'[^a-zA-Z0-9\s]'
  text = re.sub(pattern, '', text)
  return text

remove_special_characters('Sómě Áccentěd těxt')
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt_tab')

print(stopwords.words('english'))

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Sample text
text = "This is a simple example to demonstrate removing stop words using NLTK."

# Tokenize the text
words = word_tokenize(text)

# Get the list of stopwords for English
stop_words = set(stopwords.words('english'))

# Remove stopwords
filtered_words = [word for word in words if word.lower() not in stop_words] # word.lower() not in stop_words,  ensures the lowercase word isn't in the stop_words collection.

print("Original Text:", text)
print("Filtered Text:", " ".join(filtered_words))
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
def lemmatize_text(text):
  lemmatizer = WordNetLemmatizer()
  return ' '.join([lemmatizer.lemmatize(word) for word in text.split()])

lemmatize_text("The striped bats are hanging on their feet for best")
lemmatize_text("My system keeps crashing, his crashed yesterday, ours crashes daily")
```

```python
from nltk.stem import WordNetLemmatizer

def simple_stemmer(text):
  ps = nltk.porter.PorterStemmer()
  text = ' '.join([ps.stem(word) for word in text.split()])
  return text

simple_stemmer("My system keeps crashing, his crashed yesterday, ours crashes daily")
```

```python
#Install Spacy
!pip install spacy --quiet
!python -m spacy download en_core_web_sm
import spacy
nlp=spacy.load("en_core_web_sm")
```

```python
# Example text
text="Barack Obama is an American politician who served as the 44th President of the United States from 2009 to 2017. He is the first African American to have served as president, as well as the first born outside the contiguous United States."

doc3=nlp(text)
from spacy import displacy
displacy.render(doc3, style='ent', jupyter=True)
# Extract entities
entities = [(ent.text, ent.label_) for ent in doc3.ents]
print(entities)
```

```python
!pip install pandas
import pandas as pd
# Example: Creating a DataFrame
data = []
for ent in doc3.ents:
    data.append({"Entity": ent.text, "Label": ent.label_})
df = pd.DataFrame(data)
print(df)

#Process a sentence
doc = nlp('She saw a bear. She was very afraid.')

type(doc)
#Print POS tags for each word 2 for word in doc:
for word in doc:
  print (word, word.pos_)
```
