## 1. Handling Upper case

```python
def to_lowercase(text):
    """
    Convert text to lowercase as part of text normalization
    """
    return text.lower()

# Example usage
to_lowercase("<html> Natural Language Processing is AWESOME! </html>")
```

## 2.Handling HTML Text
### We can use Beautiful Soup package to clean Web data

```python
from bs4 import BeautifulSoup
```

```python
def strip_html_tags(text):
    soup = BeautifulSoup(text, "html.parser")
    stripped_text = soup.get_text()
    return stripped_text
```

```python
strip_html_tags("<html><h2>Some important text</h2></html>")
```

## 3. Remove special characters

```python
import re
```

```python
def remove_special_characters (text, remove_digits=False):
  #Using regex
  if remove_digits:
    pattern = r'[^a-zA-Z\s]'
  else:
    pattern = r'[^a-zA-Z0-9\s]'
  text = re.sub(pattern, '', text)
  return text
```

```python
remove_special_characters("Well this was fun! What do you think? 123#@!",
                         remove_digits=False)
```

```python
remove_special_characters("Well this was fun! What do you think? 123#@!",
                         remove_digits=True)
```

```python
remove_special_characters('Sómě Áccentěd těxt')
```

## 5. Removing the Stop Words

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
print(stopwords.words('english'))
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Sample text
text = "This is a simple example to demonstrate removing stop words using NLTK."

# Tokenize the text
words = word_tokenize(text)

# Get the list of stopwords for English
stop_words = set(stopwords.words('english'))

# Remove stopwords
filtered_words = [word for word in words if word.lower() not in stop_words] # word.lower() not in stop_words,  ensures the lowercase word isn't in the stop_words collection.

print("Original Text:", text)
print("Filtered Text:", " ".join(filtered_words))
```

## 6. Text Lemmatization

* **Lemmatization**

   - Reduces words to their dictionary form using vocabulary and morphological analysis. Preserves more meaning than stemming.

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
def lemmatize_text(text):
  lemmatizer = WordNetLemmatizer()
  return ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
lemmatize_text("My system keeps crashing, his crashed yesterday, ours crashes daily")
```

## 7. Text Stemming

```python
def simple_stemmer(text):
  ps = nltk.porter.PorterStemmer()
  text = ' '.join([ps.stem(word) for word in text.split()])
  return text
```

```python
simple_stemmer("My system keeps crashing, his crashed yesterday, ours crashes daily")
```

```python
stemmer = nltk.porter.PorterStemmer()
words = ["running", "runs", "runner"]
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```
