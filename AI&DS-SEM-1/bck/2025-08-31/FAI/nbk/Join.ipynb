```python
import pandas as pd
df2 = pd.DataFrame({
    'Employee_ID': [101,103,104,105,106],
    'Project_name':['A','B','C','D','E']
})
df2
```

Inner Join

```python
import pandas as pd
results = pd.merge(df1,df2, on ='Employee_ID', how='inner')
results
```

Outer Join

```python
import pandas as pd
results = pd.merge(df1,df2, on ='Employee_ID', how='outer')
results
```

Left Join

```python
import pandas as pd
results = pd.merge(df1,df2, on ='Employee_ID', how='left')
results
```

```python
import pandas as pd
results = pd.merge(df1,df2, on ='Employee_ID', how='right')
results
```
