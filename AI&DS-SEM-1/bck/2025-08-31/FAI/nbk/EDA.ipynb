## Why isna().count() doesn't change after fillna()

**The Issue:** `isna().count()` always returns the **total number of rows** (including NaN values), not the number of missing values.

**Explanation:**
- `isna()` returns a boolean Series (True for missing, False for non-missing)
- `count()` counts **all non-NaN values** in the Series
- Since boolean values (True/False) are never NaN, `count()` always returns the total length

**What you should use instead:**
- `isna().sum()` - counts True values (number of missing values)
- `notna().sum()` - counts False values (number of non-missing values)
- `isna().mean()` - percentage of missing values

````python
# Demonstration of the difference

# First, let's create a sample with missing values to demonstrate
sample_ages = pd.Series([25, 30, np.nan, 35, np.nan, 40])
print("Sample with missing values:")
print(f"Data: {sample_ages.tolist()}")
print()

# WRONG WAY - using count() after isna()
print("❌ WRONG: Using isna().count()")
print(f"Before fillna: {sample_ages.isna().count()}")
filled_ages = sample_ages.fillna(sample_ages.mean())
print(f"After fillna:  {filled_ages.isna().count()}")
print("→ Count doesn't change because count() counts all boolean values!")
print()

# RIGHT WAY - using sum() after isna()
print("✅ CORRECT: Using isna().sum()")
print(f"Before fillna: {sample_ages.isna().sum()}")
print(f"After fillna:  {filled_ages.isna().sum()}")
print("→ Sum counts True values (missing values)")
print()

print("Other useful methods:")
print(f"Total non-missing before: {sample_ages.notna().sum()}")
print(f"Total non-missing after:  {filled_ages.notna().sum()}")
print(f"Percentage missing before: {sample_ages.isna().mean():.2%}")
print(f"Percentage missing after:  {filled_ages.isna().mean():.2%}")
````

````python
# Fix for your original code - use sum() instead of count()

print("Checking missing values in Age column (CORRECTED):")
print(f"Missing values before fillna: {df['Age'].isna().sum()}")

# Fill missing values with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())

print(f"Missing values after fillna:  {df['Age'].isna().sum()}")
print(f"Total rows in dataset: {len(df)}")
print(f"Non-missing Age values: {df['Age'].notna().sum()}")
````

````python
# 🎨 COMPLETE PARAMETER EXAMPLE: Professional Publication-Quality Plot

# Step 1: Create figure with specific dimensions and DPI
plt.figure(figsize=(10, 6),    # Larger size for detailed analysis
          dpi=100,             # Standard resolution for screen display
          facecolor='white',   # Clean white background
          edgecolor='black')   # Professional black border

# Step 2: Create sophisticated bar plot with multiple parameters
sns.barplot(x='Pclass',              # x-axis: Passenger class
           y='Fare',                 # y-axis: Average fare by class
           data=df,                  # Data source
           palette='viridis',        # Professional color gradient
           edgecolor='black',        # Black borders for clarity
           linewidth=1.2,            # Medium border thickness
           alpha=0.8)                # Slight transparency for elegance

# Step 3: Add comprehensive styling
plt.title("Average Fare by Passenger Class", 
         fontsize=16,              # Large, readable title
         fontweight='bold',        # Bold for emphasis
         pad=20,                   # Extra spacing from plot
         color='darkblue')         # Professional color

plt.xlabel("Passenger Class", 
          fontsize=12,             # Medium size for readability
          fontweight='semibold',   # Slightly bold
          labelpad=10,             # Spacing from axis
          color='black')           # Standard black

plt.ylabel("Average Fare ($)", 
          fontsize=12,             # Consistent with x-label
          fontweight='semibold',   # Matching weight
          labelpad=10,             # Consistent spacing
          color='black')           # Standard color

# Step 4: Professional layout and display
plt.tight_layout()                # Optimize spacing automatically
plt.grid(True,                    # Add grid for easier reading
        alpha=0.3,                # Subtle grid lines
        linestyle='--',           # Dashed style
        linewidth=0.5)            # Thin lines

plt.show()                        # Display the final result

# 💡 WHAT MAKES THIS PROFESSIONAL:
print("✨ Professional Design Elements:")
print("• Consistent font sizes (16 for title, 12 for labels)")
print("• Color harmony (darkblue title, black labels, viridis bars)")
print("• Proper spacing (pad=20 for title, labelpad=10 for axes)")
print("• Visual hierarchy (bold title, semibold labels)")
print("• Clean presentation (tight_layout, subtle grid, alpha transparency)")

# 📊 PARAMETER IMPACT ANALYSIS:
print("\n📊 How Each Parameter Enhances the Visualization:")
print("• figsize=(10,6): Provides space for detailed analysis")
print("• palette='viridis': Scientifically accurate, colorblind-friendly")
print("• edgecolor='black': Creates clear bar separation")
print("• alpha=0.8: Adds sophistication without losing readability")
print("• grid with alpha=0.3: Aids reading without overwhelming the data")
````

### **9.1 Comprehensive Descriptive Statistics**

Beyond basic `describe()`, we implement a custom function for detailed statistical analysis:

**Enhanced Statistics Include:**
- **Central Tendency**: Mean, Median, Mode
- **Variability**: Standard Deviation, Variance, Range
- **Distribution Shape**: Skewness, Kurtosis
- **Data Quality**: Count, Missing values
- **Error Measures**: Standard Error

**Statistical Interpretations:**
- **Skewness**: Measures asymmetry (-1 to 1 range)
  - Negative: Left-skewed (tail extends left)
  - Positive: Right-skewed (tail extends right)
  - Zero: Symmetric distribution
- **Kurtosis**: Measures tail heaviness
  - High: Heavy tails, more outliers
  - Low: Light tails, fewer outliers

````python
import numpy as np
from scipy import stats
import pandas as pd

def descriptive_stats_table(data):
    stats_dict = {
        'Mean': np.mean(data),
        'Standard Error': stats.sem(data),
        'Median': np.median(data),
        'Mode': stats.mode(data, nan_policy="omit").mode.item(),
        'Standard Deviation': np.std(data, ddof=1),
        'Sample Variance': np.var(data, ddof=1),
        'Kurtosis': stats.kurtosis(data, nan_policy="omit"),
        'Skewness': stats.skew(data, nan_policy="omit"),
        'Range': np.ptp(data),
        'Minimum': np.min(data),
        'Maximum': np.max(data),
        'Sum': np.sum(data),
        'Count': len(data)
    }
    df = pd.DataFrame(list(stats_dict.items()), columns=['Parameter', 'Value'])
    return df

# Example usage
data = [1, 2, 3, 4, 5, 5, 6, 7, 8, 9]  # Replace with your dataset
result_table = descriptive_stats_table(data)
print(result_table)
````

## 🎯 **1. Introduction to Exploratory Data Analysis** {#introduction}

### **What is EDA?**
Exploratory Data Analysis (EDA) is an approach to analyzing datasets to summarize their main characteristics, often using statistical graphics and other data visualization methods. It was pioneered by statistician John Tukey in the 1970s.

### **Core Philosophy of EDA**
> *"The greatest value of a picture is when it forces us to notice what we never expected to see."* - John Tukey

### **Primary Objectives:**
1. **Understand the data structure** and composition
2. **Identify patterns, trends, and relationships**
3. **Detect outliers and anomalies**
4. **Test underlying assumptions**
5. **Generate hypotheses** for further investigation
6. **Guide feature selection** and engineering decisions

### **EDA vs. Traditional Statistical Analysis**
| **Aspect** | **EDA** | **Traditional Analysis** |
|------------|---------|-------------------------|
| **Approach** | Exploratory, hypothesis-generating | Confirmatory, hypothesis-testing |
| **Focus** | Discovery and understanding | Validation and inference |
| **Methods** | Visual, interactive, flexible | Mathematical, rigid procedures |
| **Assumptions** | Minimal prior assumptions | Strong distributional assumptions |

In our Titanic analysis, EDA revealed critical insights:
- **Gender-based survival patterns** (women had higher survival rates)
- **Class-based disparities** (first-class passengers survived more)

## 📈 **4. Statistical Foundations of EDA** {#statistics}

### **4.1 Descriptive Statistics**

**Central Tendency Measures:**

1. **Mean (μ)**
   - **Formula**: μ = Σx / n
   - **Use case**: Best for symmetric distributions
   - **Titanic example**: Average age of passengers

2. **Median**
   - **Definition**: Middle value when data is sorted
   - **Advantage**: Robust to outliers
   - **Titanic example**: Median fare paid

3. **Mode**
   - **Definition**: Most frequently occurring value
   - **Application**: Best for categorical data
   - **Titanic example**: Most common embarkation port

**Dispersion Measures:**

1. **Standard Deviation (σ)**
   - **Formula**: σ = √(Σ(x-μ)² / n)
   - **Interpretation**: Average distance from mean
   - **Titanic insight**: Age variability across passengers

2. **Variance (σ²)**
   - **Formula**: σ² = Σ(x-μ)² / n
   - **Use**: Measure of spread
   - **Application**: Fare price variability

3. **Range**
   - **Formula**: Range = Max - Min
   - **Limitation**: Sensitive to outliers
   - **Example**: Age range from 0.42 to 80 years

4. **Interquartile Range (IQR)**
   - **Formula**: IQR = Q3 - Q1
   - **Advantage**: Robust to outliers
   - **Use**: Box plot construction

### **4.2 Distribution Analysis**

**Skewness:**
- **Right-skewed (positive)**: Tail extends to the right
- **Left-skewed (negative)**: Tail extends to the left
- **Interpretation**: 
  - Skewness > 1: Highly right-skewed
  - -1 < Skewness < 1: Approximately symmetric
  - Skewness < -1: Highly left-skewed

**Kurtosis:**
- **Mesokurtic**: Normal distribution (kurtosis ≈ 3)
- **Leptokurtic**: Heavy tails (kurtosis > 3)
- **Platykurtic**: Light tails (kurtosis < 3)

**Titanic Distribution Examples:**
```python
from scipy import stats

# Age distribution analysis
age_skewness = stats.skew(df['Age'].dropna())
age_kurtosis = stats.kurtosis(df['Age'].dropna())

print(f"Age Skewness: {age_skewness:.3f}")
print(f"Age Kurtosis: {age_kurtosis:.3f}")
```

### **4.3 Probability Distributions**

**Common Distributions in EDA:**

1. **Normal Distribution**
   - **Characteristics**: Bell-shaped, symmetric
   - **Parameters**: Mean (μ) and standard deviation (σ)
   - **68-95-99.7 Rule**: Empirical rule for data spread

2. **Binomial Distribution**
   - **Use case**: Binary outcomes (survived/died)
   - **Parameters**: n (trials) and p (probability)
   - **Titanic application**: Survival probability

3. **Poisson Distribution**
   - **Use case**: Count data (family size)
   - **Parameter**: λ (rate parameter)
   - **Application**: Number of family members

### **4.4 Hypothesis Testing in EDA**

**Common Tests:**

1. **Chi-Square Test**
   - **Purpose**: Test independence of categorical variables
   - **Titanic application**: Gender and survival independence
   - **Null hypothesis**: Variables are independent

2. **T-Test**
   - **Purpose**: Compare means between groups
   - **Application**: Age difference between survivors and non-survivors
   - **Types**: One-sample, two-sample, paired

3. **ANOVA (Analysis of Variance)**
   - **Purpose**: Compare means across multiple groups
   - **Application**: Fare differences across passenger classes
   - **Extension**: Post-hoc tests for pairwise comparisons

### **5.1 Missing Data Analysis**

**Types of Missingness:**

1. **Missing Completely at Random (MCAR)**
   - **Definition**: Missingness is unrelated to any variables
   - **Test**: Little's MCAR test
   - **Implication**: Safe to delete or impute randomly

2. **Missing at Random (MAR)**
   - **Definition**: Missingness depends on observed variables
   - **Example**: Age missing more for certain passenger classes
   - **Strategy**: Use observed variables for imputation

3. **Missing Not at Random (MNAR)**
   - **Definition**: Missingness depends on unobserved values
   - **Example**: High-fare passengers not reporting cabin numbers
   - **Challenge**: Requires domain knowledge to handle

**Missing Data Patterns:**

### **5.2 Outlier Detection**

**Statistical Methods:**

1. **Z-Score Method**
   - **Formula**: z = (x - μ) / σ
   - **Threshold**: |z| > 3 (or 2.5)
   - **Assumption**: Normal distribution

2. **Interquartile Range (IQR) Method**
   - **Formula**: Outliers < Q1 - 1.5×IQR or > Q3 + 1.5×IQR
   - **Advantage**: Distribution-free
   - **Application**: Box plot construction

3. **Modified Z-Score**
   - **Uses median**: More robust than standard Z-score
   - **Formula**: M = 0.6745 × (x - median) / MAD
   - **Threshold**: |M| > 3.5

**Visual Methods:**

```python
# Bootstrap confidence intervals
def bootstrap_mean(data, n_bootstrap=1000, confidence=0.95):
    bootstrap_means = []
    for _ in range(n_bootstrap):
        sample = np.random.choice(data, size=len(data), replace=True)
        bootstrap_means.append(np.mean(sample))
    
    alpha = 1 - confidence
    lower = np.percentile(bootstrap_means, 100 * alpha/2)
    upper = np.percentile(bootstrap_means, 100 * (1 - alpha/2))
    
    return lower, upper

# Example: Age confidence interval
age_ci = bootstrap_mean(df['Age'].dropna())
print(f"Age mean 95% CI: [{age_ci[0]:.2f}, {age_ci[1]:.2f}]")
```

## ⚠️ **9. Common Pitfalls & Solutions** {#pitfalls}

### **9.1 Data-Related Pitfalls**

**Pitfall 1: Ignoring Missing Data Patterns**
- **Problem**: Treating all missing data the same way
- **Solution**: Analyze missingness patterns (MCAR, MAR, MNAR)
- **Titanic example**: Age missing more in 3rd class passengers

**Pitfall 2: Outlier Mishandling**
- **Problem**: Automatically removing all outliers
- **Solution**: Investigate outliers first - they might be valuable
- **Example**: High-fare passengers might represent important VIPs

**Pitfall 3: Data Leakage**
- **Problem**: Using future information to predict past events
- **Solution**: Temporal validation of features
- **Prevention**: Always consider the data collection process

### **9.2 Statistical Pitfalls**

**Pitfall 4: Multiple Testing Problem**
- **Problem**: Testing many hypotheses without adjustment
- **Solution**: Bonferroni correction or FDR control
```python
from statsmodels.stats.multitest import multipletests

# Multiple hypothesis testing correction
p_values = [0.01, 0.03, 0.05, 0.07, 0.1]  # Example p-values
rejected, corrected_p, _, _ = multipletests(p_values, method='bonferroni')
print(f"Corrected p-values: {corrected_p}")
```

**Pitfall 5: Simpson's Paradox**
- **Problem**: Trend reverses when data is aggregated
- **Solution**: Always examine subgroups separately
- **Titanic example**: Overall survival rate vs. class-specific rates

**Pitfall 6: Survivorship Bias**
- **Problem**: Only analyzing surviving/successful cases
- **Solution**: Ensure representative sampling
- **Example**: Only analyzing passengers who boarded (missing those who didn't)

### **9.3 Visualization Pitfalls**

**Pitfall 7: Misleading Scales**
- **Problem**: Non-zero baselines, inappropriate scales
- **Solution**: Always start from zero for bar charts
```python
# Good practice: explicit y-axis limits
plt.bar(categories, values)
plt.ylim(0, max(values) * 1.1)  # Start from zero
```

**Pitfall 8: Overplotting**
- **Problem**: Too many points obscuring patterns
- **Solution**: Use transparency, sampling, or binning
```python
plt.scatter(x, y, alpha=0.3)  # Transparency
# or use hexbin for dense data
plt.hexbin(x, y, gridsize=20)
```

### **9.4 Interpretation Pitfalls**

**Pitfall 9: Cherry-picking Results**
- **Problem**: Only reporting favorable findings
- **Solution**: Report all analyses conducted
- **Practice**: Pre-register analysis plans

**Pitfall 10: Causal Inference from Correlation**
- **Problem**: Assuming correlation implies causation
- **Solution**: Use causal inference methods or acknowledge limitations
- **Example**: Age and fare correlation doesn't mean age causes higher fares

```python
from scipy.stats import chi2_contingency

# Test independence of gender and survival
contingency_table = pd.crosstab(df['Sex'], df['Survived'])
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of freedom: {dof}")

# Effect size (Cramér's V)
n = contingency_table.sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(contingency_table.shape) - 1)))
```
