```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0,6])
y = np.array([0,100])
plt.plot(x, y,                    # Data arrays (x and y coordinates)
         color='blue',             # Line color: 'red', '#FF5733', (0.1, 0.2, 0.3)
         linestyle='-',            # Line style: '-', '--', '-.', ':', 'None'
         linewidth=2,              # Line thickness: float value
         marker='o',               # Point markers: 'o', 's', '^', '*', '+', 'x'
         markersize=8,             # Marker size: float value
         markerfacecolor='red',    # Marker fill color
         markeredgecolor='black',  # Marker border color
         markeredgewidth=1,        # Marker border thickness
         alpha=0.7,                # Transparency: 0.0 (transparent) to 1.0 (opaque)
         label='Data Series',      # Legend label
         zorder=1)                 # Drawing layer (higher = front)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

sales = [3000, 1200, 5000, 1800, 4500]
mobile = ('Vivo', 'Oppo', 'Micromax', 'Samsung', 'Apple')

y_pos = np.arange(len(mobile))

plt.bar(y_pos, sales)

# plt.bar() function parameters:
# plt.bar(x, height, width=0.8, bottom=None, align='center', data=None, **kwargs)
# x: x-coordinates of bars (sequence of scalars)
# height: heights of bars (sequence of scalars)
# width: width of bars (float or array-like, default 0.8)
# bottom: y-coordinate of bar bottoms (float or array-like, default None)
# align: alignment of bars ('center' or 'edge', default 'center')
# color: face color of bars (color or sequence of colors)
# edgecolor: edge color of bars (color or sequence of colors)
# linewidth: width of bar edges (float, default None)
# alpha: transparency level (float between 0-1, default None)
# label: label for legend (string, default None)
# hatch: hatching pattern ('/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*')
# capsize: length of error bar caps (float, default None)
# error_kw: dictionary of error bar properties
# log: use log scale on y-axis (bool, default False)

plt.xticks(y_pos, mobile)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
python = (50, 65, 40, 35, 77)
machinelearning = (55, 72, 94, 70, 85)

ind = np.arange(N)
width = 0.35

plt.bar(ind, python, width, label='Python')
plt.bar(ind + width, machinelearning, width, label='Machine Learning')

plt.ylabel('Scores')
plt.title('Scores by Student')

plt.xticks(ind + width / 2, ('A', 'B', 'C', 'D', 'E'))
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = 'Python', 'Machine Learning', 'R', 'Big Data'
sizes = [85, 90, 85, 70]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# plt.pie() function parameters:
# plt.pie(x, explode=None, labels=None, colors=None, autopct=None,
#         pctdistance=0.6, shadow=False, labeldistance=1.1,
#         startangle=None, radius=None, counterclock=True,
#         wedgeprops=None, textprops=None, center=(0, 0), frame=False       
# )
# where,
# x: sizes of pie slices (sequence of scalars)
# explode: fraction to offset each slice (sequence of floats, default None)
# labels: labels for each slice (sequence of strings, default None)
# colors: colors for each slice (sequence of color specifications, default None)
# autopct: format string for slice percentages (string or function, default None)
# pctdistance: ratio of pie radius to position of percentage labels (float, default 0.6)
# shadow: draw a shadow beneath the pie (bool, default False)
# labeldistance: radial distance for slice labels (float, default 1.1)
# startangle: starting angle of pie (float, default None)
# radius: radius of pie (float, default None)
# counterclock: draw slices counterclockwise (bool, default True)
# wedgeprops: dictionary of wedge properties (default None)
# textprops: dictionary of text properties (default None)
# center: center of pie (x, y) (tuple, default (0, 0))
# frame: draw a frame around the pie (bool, default False)

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, shadow=True, startangle=140)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = 'Python', 'Machine Learning', 'R', 'Big Data'
sizes = [85, 90, 85, 70]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0.2, 0, 0)

plt.pie(sizes, labels=labels, explode=explode, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title('Skill Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = 'Python', 'Machine Learning', 'R', 'Big Data'
sizes = [85, 90, 85, 70]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    shadow=True,
    startangle=90
)

# Create a white circle in the center to make it a donut chart
center_circle = plt.Circle((0, 0), 0.5, color='white', fc='white', linewidth=0)
fig = plt.gcf()  # Get current figure
fig.gca().add_artist(center_circle)  # Add the center circle to current axes

# gcf() gets the current figure (the entire plot window)
# gca() gets the current axes (the plotting area within the figure)

plt.axis('equal')
plt.title('Skill Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df_iris = pd.read_csv("iris.csv")
df_iris['sepal.width'].plot.hist(bins=15,edgecolor='black',alpha=0.7, color='skyblue')

plt.title('Distribution of Sepal Width (cm)', fontsize=14, pad=20)
plt.xlabel('Sepal Width (cm)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.boxplot(df_iris['sepal.width'])
plt.title('Sepal Width Distribution')
plt.ylabel('Width (cm)')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.boxplot(df_iris['sepal.width'])
# Add text annotations to label the quartiles and extremes of the boxplot
# plt.text(x, y, s) adds text at position (x, y) with string s
# x=1.1 positions the text slightly to the right of the boxplot (which is at x=1)
# y-coordinates correspond to the statistical values of the sepal width data

plt.text(x=1.1, y=df_iris['sepal.width'].min(), s='min')  # Label the minimum value
plt.text(x=1.1, y=df_iris['sepal.width'].quantile(0.25), s='Q1')  # Label the first quartile (25th percentile)
plt.text(x=1.1, y=df_iris['sepal.width'].median(), s='median (Q2)')  # Label the median (50th percentile)
plt.text(x=1.1, y=df_iris['sepal.width'].quantile(0.75), s='Q3')  # Label the third quartile (75th percentile)
plt.text(x=1.1, y=df_iris['sepal.width'].max(), s='max')  # Label the maximum value

plt.show()  # Display the plot with all annotations
```
