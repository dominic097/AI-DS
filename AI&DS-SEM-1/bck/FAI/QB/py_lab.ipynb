```python
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05',
             '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-09', '2023-01-10'],
    'Store': ['Store A', 'Store A', 'Store B', 'Store B', 'Store A',
              'Store C', 'Store C', 'Store A', 'Store B', 'Store C'],
    'Product': ['Widget A', 'Widget B', 'Widget A', 'Widget C', np.nan,
                'Widget A', 'Widget B', 'Widget C', 'Widget B', 'Widget C'],
    'Sales': [200, 150, 300, np.nan, 250, 400, 100, 150, np.nan, 350],
    'Quantity': [10, 5, 15, 8, 20, 20, 2, 6, 5, 10]
}


df = pd.DataFrame(data)

print(df)

print("Handling Missing Values:")
print(df.isna().sum())

missing_columns = df.columns[df.isnull().any()].tolist()

print("Columns with missing values:", missing_columns)

print("Filling missing values: Product with 'Unknown', Sales with mean sales")
df["Product"].fillna("Unknown", inplace=True)
df["Sales"].fillna(df["Sales"].mean(), inplace=True)

print(df.isna().sum())

print("Date Time to datetime object")

df["Date"] = pd.to_datetime(df["Date"])

print("Group by store and product, then aggregate sales and quantity")

gdf = df.groupby(["Store", "Product"]).agg({"Sales": "sum", "Quantity": "sum"}).reset_index()

gdf.describe()

```
