# 📊 Covariance and Correlation: Step-by-Step Manual Calculation

## 🧊 Ice Cream Café Problem

**Scenario:** A café owner wants to know if higher temperatures increase ice cream sales. He records data over 6 days.

| Day | Temperature (°C) | Ice Cream Sales (units) |
|-----|------------------|-------------------------|
| 1   | 20               | 40                      |
| 2   | 22               | 45                      |
| 3   | 25               | 60                      |
| 4   | 27               | 65                      |
| 5   | 30               | 70                      |
| 6   | 32               | 85                      |

**Research Question:** Is there a relationship between temperature and ice cream sales?

## 📐 Mathematical Formulas

### **Population Covariance:**
```
σ(X,Y) = Σ[(Xi - μX)(Yi - μY)] / N
```

### **Sample Covariance:**
```
s(X,Y) = Σ[(Xi - X̄)(Yi - Ȳ)] / (n-1)
```

### **Correlation Coefficient:**
```
r = Cov(X,Y) / (σX × σY)
```

Where:
- **X** = Temperature, **Y** = Ice Cream Sales
- **μX, μY** = Population means
- **X̄, Ȳ** = Sample means
- **σX, σY** = Standard deviations
- **N** = Population size, **n** = Sample size

## 🔢 Step 1: Data Setup and Mean Calculation

### **Given Data:**
- **X (Temperature):** [20, 22, 25, 27, 30, 32]
- **Y (Ice Cream Sales):** [40, 45, 60, 65, 70, 85]
- **Sample size (n):** 6

### **Calculate Means:**

**Mean of X (Temperature):**
```
X̄ = (20 + 22 + 25 + 27 + 30 + 32) / 6
X̄ = 156 / 6
X̄ = 26°C
```

**Mean of Y (Ice Cream Sales):**
```
Ȳ = (40 + 45 + 60 + 65 + 70 + 85) / 6
Ȳ = 365 / 6
Ȳ = 60.83 units
```

### **Summary:**
- **Mean Temperature (X̄):** 26°C
- **Mean Ice Cream Sales (Ȳ):** 60.83 units

## 📏 Step 2: Calculate Deviations from Mean

### **Deviation Formula:**
- **Xi - X̄** = Each temperature minus mean temperature
- **Yi - Ȳ** = Each sales value minus mean sales

### **Detailed Calculations:**

| Day | Xi | Yi | (Xi - X̄) | (Yi - Ȳ) | Calculation |
|-----|----|----|-----------|-----------|-------------|
| 1 | 20 | 40 | 20 - 26 = **-6** | 40 - 60.83 = **-20.83** | Below average temp, below average sales |
| 2 | 22 | 45 | 22 - 26 = **-4** | 45 - 60.83 = **-15.83** | Below average temp, below average sales |
| 3 | 25 | 60 | 25 - 26 = **-1** | 60 - 60.83 = **-0.83** | Slightly below average temp, slightly below average sales |
| 4 | 27 | 65 | 27 - 26 = **+1** | 65 - 60.83 = **+4.17** | Above average temp, above average sales |
| 5 | 30 | 70 | 30 - 26 = **+4** | 70 - 60.83 = **+9.17** | Above average temp, above average sales |
| 6 | 32 | 85 | 32 - 26 = **+6** | 85 - 60.83 = **+24.17** | Above average temp, above average sales |

### **Key Observations:**
- **Negative deviations:** Below average values
- **Positive deviations:** Above average values
- **Pattern:** When temperature is above average, sales are also above average (positive relationship!)

### **Verification (Sum of Deviations = 0):**
- **Sum of (Xi - X̄):** (-6) + (-4) + (-1) + (+1) + (+4) + (+6) = **0** ✅
- **Sum of (Yi - Ȳ):** (-20.83) + (-15.83) + (-0.83) + (+4.17) + (+9.17) + (+24.17) = **0** ✅

## 🔄 Step 3: Calculate Products of Deviations

### **Product Formula:**
**(Xi - X̄) × (Yi - Ȳ)** = Multiply temperature deviation by sales deviation

### **Detailed Product Calculations:**

| Day | (Xi - X̄) | (Yi - Ȳ) | (Xi - X̄) × (Yi - Ȳ) | Calculation | Interpretation |
|-----|-----------|-----------|----------------------|-------------|----------------|
| 1 | -6 | -20.83 | (-6) × (-20.83) = **+125.00** | Negative × Negative = Positive | Both below average → Positive contribution |
| 2 | -4 | -15.83 | (-4) × (-15.83) = **+63.33** | Negative × Negative = Positive | Both below average → Positive contribution |
| 3 | -1 | -0.83 | (-1) × (-0.83) = **+0.83** | Negative × Negative = Positive | Both below average → Positive contribution |
| 4 | +1 | +4.17 | (+1) × (+4.17) = **+4.17** | Positive × Positive = Positive | Both above average → Positive contribution |
| 5 | +4 | +9.17 | (+4) × (+9.17) = **+36.67** | Positive × Positive = Positive | Both above average → Positive contribution |
| 6 | +6 | +24.17 | (+6) × (+24.17) = **+145.00** | Positive × Positive = Positive | Both above average → Positive contribution |

### **Sum of Products:**
```
Σ[(Xi - X̄)(Yi - Ȳ)] = 125.00 + 63.33 + 0.83 + 4.17 + 36.67 + 145.00
Σ[(Xi - X̄)(Yi - Ȳ)] = 375.00
```

### **💡 Key Insight:**
**All products are positive!** This indicates a **positive relationship** between temperature and ice cream sales.

- When both variables are **below average** → Product is **positive**
- When both variables are **above average** → Product is **positive**
- This pattern suggests that as temperature increases, ice cream sales also increase!

## 🧮 Step 4: Calculate Covariance

### **We calculated:** Σ[(Xi - X̄)(Yi - Ȳ)] = 375.00

Now we divide by appropriate denominator:

### **📊 Sample Covariance (Most Common):**
```
Sample Covariance = Σ[(Xi - X̄)(Yi - Ȳ)] / (n-1)
Sample Covariance = 375.00 / (6-1)
Sample Covariance = 375.00 / 5
Sample Covariance = 75.00
```

### **🏛️ Population Covariance:**
```
Population Covariance = Σ[(Xi - X̄)(Yi - Ȳ)] / N
Population Covariance = 375.00 / 6
Population Covariance = 62.50
```

### **⚖️ Comparison:**
- **Sample Covariance:** 75.00 (uses n-1 = 5)
- **Population Covariance:** 62.50 (uses N = 6)
- **Difference:** Sample covariance is higher due to Bessel's correction

### **💡 Interpretation:**
- **Positive covariance (75.00)** confirms our observation
- Temperature and ice cream sales **move together** in the **same direction**
- When temperature increases, ice cream sales tend to increase too!

## 📐 Step 5: Calculate Standard Deviations (for Correlation)

### **Standard Deviation Formula:**
```
s = √[Σ(Xi - X̄)² / (n-1)]
```

### **📊 Standard Deviation of Temperature (X):**

**Calculate squared deviations:**
| Day | (Xi - X̄) | (Xi - X̄)² |
|-----|-----------|-----------|
| 1 | -6 | (-6)² = 36 |
| 2 | -4 | (-4)² = 16 |
| 3 | -1 | (-1)² = 1 |
| 4 | +1 | (+1)² = 1 |
| 5 | +4 | (+4)² = 16 |
| 6 | +6 | (+6)² = 36 |

**Sum:** Σ(Xi - X̄)² = 36 + 16 + 1 + 1 + 16 + 36 = **106**

**Standard deviation:**
```
sX = √[106 / (6-1)] = √[106 / 5] = √21.2 = 4.61°C
```

### **🧊 Standard Deviation of Ice Cream Sales (Y):**

**Calculate squared deviations:**
| Day | (Yi - Ȳ) | (Yi - Ȳ)² |
|-----|-----------|-----------|
| 1 | -20.83 | (-20.83)² = 433.89 |
| 2 | -15.83 | (-15.83)² = 250.59 |
| 3 | -0.83 | (-0.83)² = 0.69 |
| 4 | +4.17 | (+4.17)² = 17.39 |
| 5 | +9.17 | (+9.17)² = 84.09 |
| 6 | +24.17 | (+24.17)² = 584.19 |

**Sum:** Σ(Yi - Ȳ)² = 433.89 + 250.59 + 0.69 + 17.39 + 84.09 + 584.19 = **1370.84**

**Standard deviation:**
```
sY = √[1370.84 / (6-1)] = √[1370.84 / 5] = √274.17 = 16.56 units
```

### **📋 Summary:**
- **Standard deviation of Temperature (sX):** 4.61°C
- **Standard deviation of Ice Cream Sales (sY):** 16.56 units

## 🎯 Step 6: Calculate Correlation Coefficient

### **Correlation Formula:**
```
r = Cov(X,Y) / (sX × sY)
```

### **Using our calculated values:**
- **Sample Covariance:** 75.00
- **Standard deviation of Temperature (sX):** 4.61°C  
- **Standard deviation of Ice Cream Sales (sY):** 16.56 units

### **Final Calculation:**
```
r = 75.00 / (4.61 × 16.56)
r = 75.00 / 76.34
r = 0.982
```

## 📊 Results Summary

| Measure | Value | Interpretation |
|---------|-------|----------------|
| **Sample Covariance** | 75.00 | Positive relationship |
| **Population Covariance** | 62.50 | Positive relationship |
| **Correlation Coefficient** | 0.982 | **Very Strong Positive** |

## 💡 Interpretation

### **Correlation Strength Scale:**
- **0.90 to 1.00:** Very Strong ← **Our result: 0.982**
- **0.70 to 0.89:** Strong
- **0.50 to 0.69:** Moderate
- **0.30 to 0.49:** Weak
- **0.00 to 0.29:** Very Weak

### **What This Means:**
1. **Direction:** Positive (as temperature increases, sales increase)
2. **Strength:** Very strong relationship (r = 0.982)
3. **Predictability:** Temperature is an excellent predictor of ice cream sales
4. **Business Insight:** The café owner can confidently expect higher sales on warmer days!

### **Key Advantages of Correlation over Covariance:**
- **Standardized:** Always between -1 and +1
- **Easy to interpret:** Clear strength categories
- **Unit-independent:** Can compare different relationships

# 💻 Python Implementation

Now let's verify our manual calculations using Python!

````python
# =============================================================================
# 📚 IMPORT LIBRARIES AND SET UP DATA
# =============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Ice cream café data
temperature = np.array([20, 22, 25, 27, 30, 32])  # X values (°C)
ice_cream_sales = np.array([40, 45, 60, 65, 70, 85])  # Y values (units)

# Create DataFrame for better visualization
df = pd.DataFrame({
    'Day': range(1, 7),
    'Temperature_C': temperature,
    'Ice_Cream_Sales': ice_cream_sales
})

print("🧊 Ice Cream Café Dataset")
print("=" * 40)
print(df)
print(f"\nSample size (n): {len(temperature)}")

# Display basic statistics
print(f"\n📊 Basic Statistics:")
print(f"Temperature - Mean: {np.mean(temperature):.2f}°C, Std: {np.std(temperature, ddof=1):.2f}°C")
print(f"Sales - Mean: {np.mean(ice_cream_sales):.2f} units, Std: {np.std(ice_cream_sales, ddof=1):.2f} units")
````

````python
# =============================================================================
# 🧮 MANUAL CALCULATION VERIFICATION
# =============================================================================

def manual_covariance_correlation(x, y):
    """
    Calculate covariance and correlation manually (step by step)
    """
    n = len(x)
    
    # Step 1: Calculate means
    mean_x = np.sum(x) / n
    mean_y = np.sum(y) / n
    
    print(f"📊 Step 1: Calculate Means")
    print(f"   Mean X: {mean_x:.2f}")
    print(f"   Mean Y: {mean_y:.2f}")
    
    # Step 2: Calculate deviations
    deviations_x = x - mean_x
    deviations_y = y - mean_y
    
    print(f"\n📏 Step 2: Calculate Deviations")
    print(f"   Deviations X: {deviations_x}")
    print(f"   Deviations Y: {deviations_y}")
    
    # Step 3: Calculate products of deviations
    products = deviations_x * deviations_y
    sum_products = np.sum(products)
    
    print(f"\n🔄 Step 3: Products of Deviations")
    print(f"   Products: {products}")
    print(f"   Sum of products: {sum_products:.2f}")
    
    # Step 4: Calculate covariances
    sample_cov = sum_products / (n - 1)
    population_cov = sum_products / n
    
    print(f"\n🧮 Step 4: Calculate Covariances")
    print(f"   Sample Covariance (÷{n-1}): {sample_cov:.2f}")
    print(f"   Population Covariance (÷{n}): {population_cov:.2f}")
    
    # Step 5: Calculate standard deviations
    squares_x = deviations_x ** 2
    squares_y = deviations_y ** 2
    
    std_x = np.sqrt(np.sum(squares_x) / (n - 1))
    std_y = np.sqrt(np.sum(squares_y) / (n - 1))
    
    print(f"\n📐 Step 5: Calculate Standard Deviations")
    print(f"   Std X: {std_x:.2f}")
    print(f"   Std Y: {std_y:.2f}")
    
    # Step 6: Calculate correlation
    correlation = sample_cov / (std_x * std_y)
    
    print(f"\n🎯 Step 6: Calculate Correlation")
    print(f"   Correlation: {correlation:.6f}")
    
    return sample_cov, population_cov, correlation

# Perform manual calculations
print("🔢 MANUAL CALCULATIONS")
print("=" * 50)
manual_sample_cov, manual_pop_cov, manual_corr = manual_covariance_correlation(temperature, ice_cream_sales)
````

````python
# =============================================================================
# ✅ VERIFICATION WITH LIBRARY FUNCTIONS
# =============================================================================

print("\n\n✅ VERIFICATION WITH PYTHON LIBRARIES")
print("=" * 60)

# NumPy calculations
numpy_cov_matrix = np.cov(temperature, ice_cream_sales)
numpy_sample_cov = numpy_cov_matrix[0, 1]
numpy_pop_cov = np.cov(temperature, ice_cream_sales, bias=True)[0, 1]
numpy_corr = np.corrcoef(temperature, ice_cream_sales)[0, 1]

print(f"📚 NumPy Results:")
print(f"   Sample Covariance: {numpy_sample_cov:.2f}")
print(f"   Population Covariance: {numpy_pop_cov:.2f}")
print(f"   Correlation: {numpy_corr:.6f}")

# SciPy calculations
scipy_corr, p_value = stats.pearsonr(temperature, ice_cream_sales)

print(f"\n🔬 SciPy Results:")
print(f"   Pearson Correlation: {scipy_corr:.6f}")
print(f"   P-value: {p_value:.6f}")
print(f"   Significant: {'Yes' if p_value < 0.05 else 'No'} (α = 0.05)")

# Pandas calculations
df_corr = pd.DataFrame({'Temperature': temperature, 'Sales': ice_cream_sales})
pandas_corr = df_corr.corr().iloc[0, 1]

print(f"\n🐼 Pandas Results:")
print(f"   Correlation: {pandas_corr:.6f}")

# Comparison table
comparison_data = {
    'Method': ['Manual Calculation', 'NumPy', 'SciPy', 'Pandas'],
    'Sample Covariance': [manual_sample_cov, numpy_sample_cov, 'N/A', 'N/A'],
    'Correlation': [manual_corr, numpy_corr, scipy_corr, pandas_corr]
}

comparison_df = pd.DataFrame(comparison_data)
print(f"\n📋 COMPARISON TABLE:")
print("=" * 60)
print(comparison_df)

# Check if all methods give same results
correlation_values = [manual_corr, numpy_corr, scipy_corr, pandas_corr]
print(f"\n🎯 VERIFICATION CHECK:")
print(f"   All correlation values match: {np.allclose(correlation_values, correlation_values[0])}")
print(f"   Manual vs NumPy difference: {abs(manual_corr - numpy_corr):.8f}")
````

````python
# =============================================================================
# 🔍 UNDERSTANDING COVARIANCE MATRIX STRUCTURE
# =============================================================================

print("🔍 HOW COVARIANCE FORMULA CREATES THE MATRIX")
print("=" * 60)

# Let's break down exactly how np.cov() applies the covariance formula
temperature = np.array([20, 22, 25, 27, 30, 32])
ice_cream_sales = np.array([40, 45, 60, 65, 70, 85])

print("📊 Our Data:")
print(f"   Temperature (X): {temperature}")
print(f"   Ice Cream Sales (Y): {ice_cream_sales}")

# Calculate means
mean_temp = np.mean(temperature)
mean_sales = np.mean(ice_cream_sales)

print(f"\n📈 Means:")
print(f"   Mean Temperature (X̄): {mean_temp:.2f}")
print(f"   Mean Sales (Ȳ): {mean_sales:.2f}")

print(f"\n🧮 APPLYING COVARIANCE FORMULA TO CREATE EACH MATRIX ELEMENT:")
print("=" * 70)

# Calculate each element of the covariance matrix manually
print("🔢 Matrix Position [0,0] = Var(X) = Covariance of X with itself")
print("   Formula: Σ[(Xi - X̄)(Xi - X̄)] / (n-1)")

deviations_x = temperature - mean_temp
var_x = np.sum(deviations_x * deviations_x) / (len(temperature) - 1)

print(f"   Calculation: Σ[{deviations_x} × {deviations_x}] / {len(temperature)-1}")
print(f"   Result: {var_x:.2f}")

print(f"\n🔢 Matrix Position [0,1] = Cov(X,Y) = Covariance between X and Y")
print("   Formula: Σ[(Xi - X̄)(Yi - Ȳ)] / (n-1)")

deviations_y = ice_cream_sales - mean_sales
cov_xy = np.sum(deviations_x * deviations_y) / (len(temperature) - 1)

print(f"   Calculation: Σ[{deviations_x} × {deviations_y}] / {len(temperature)-1}")
print(f"   Result: {cov_xy:.2f}")

print(f"\n🔢 Matrix Position [1,0] = Cov(Y,X) = Covariance between Y and X")
print("   Formula: Σ[(Yi - Ȳ)(Xi - X̄)] / (n-1)")
print("   Note: This is IDENTICAL to Cov(X,Y) because multiplication is commutative!")

cov_yx = np.sum(deviations_y * deviations_x) / (len(temperature) - 1)
print(f"   Result: {cov_yx:.2f} (same as [0,1])")

print(f"\n🔢 Matrix Position [1,1] = Var(Y) = Covariance of Y with itself")
print("   Formula: Σ[(Yi - Ȳ)(Yi - Ȳ)] / (n-1)")

var_y = np.sum(deviations_y * deviations_y) / (len(temperature) - 1)
print(f"   Result: {var_y:.2f}")

print(f"\n🏗️ BUILDING THE COVARIANCE MATRIX:")
print("=" * 40)

manual_cov_matrix = np.array([[var_x, cov_xy],
                              [cov_yx, var_y]])

print("Manual Matrix Construction:")
print(f"[[{var_x:7.2f}, {cov_xy:7.2f}]")
print(f" [{cov_yx:7.2f}, {var_y:7.2f}]]")

print(f"\nNumPy's np.cov() result:")
numpy_matrix = np.cov(temperature, ice_cream_sales)
print(f"[[{numpy_matrix[0,0]:7.2f}, {numpy_matrix[0,1]:7.2f}]")
print(f" [{numpy_matrix[1,0]:7.2f}, {numpy_matrix[1,1]:7.2f}]]")

print(f"\n✅ Verification: Manual vs NumPy matrices match: {np.allclose(manual_cov_matrix, numpy_matrix)}")

print(f"\n💡 KEY INSIGHTS:")
print("=" * 20)
print("1. 🎯 [0,1] and [1,0] are IDENTICAL - both represent Cov(X,Y)")
print("2. 📊 [0,0] is Var(X) - how much X varies with itself")
print("3. 📊 [1,1] is Var(Y) - how much Y varies with itself") 
print("4. 🔄 The matrix is SYMMETRIC due to covariance symmetry property")
print("5. 🎪 We extract [0,1] because it's the covariance between our two variables!")

print(f"\n🎪 WHY WE USE [0,1] INDEXING:")
print("=" * 35)
print("When we do: numpy_sample_cov = numpy_cov_matrix[0, 1]")
print("We're saying: 'Give me the covariance between variable 0 and variable 1'")
print("Variable 0 = Temperature, Variable 1 = Ice Cream Sales")
print(f"So [0,1] = Cov(Temperature, Sales) = {numpy_matrix[0,1]:.2f}")
````

````python
# =============================================================================
# 📊 VISUALIZATION AND ANALYSIS
# =============================================================================

# Create comprehensive visualization
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot 1: Scatter plot with trend line
ax1 = axes[0]
ax1.scatter(temperature, ice_cream_sales, color='red', s=100, alpha=0.8, edgecolor='black', linewidth=1.5)

# Add trend line
z = np.polyfit(temperature, ice_cream_sales, 1)
p = np.poly1d(z)
ax1.plot(temperature, p(temperature), "b--", alpha=0.8, linewidth=2, label=f'Trend Line')

# Add mean lines
ax1.axvline(np.mean(temperature), color='green', linestyle=':', alpha=0.7, label=f'Mean Temp ({np.mean(temperature):.1f}°C)')
ax1.axhline(np.mean(ice_cream_sales), color='orange', linestyle=':', alpha=0.7, label=f'Mean Sales ({np.mean(ice_cream_sales):.1f})')

ax1.set_xlabel('Temperature (°C)', fontweight='bold')
ax1.set_ylabel('Ice Cream Sales (units)', fontweight='bold')
ax1.set_title(f'🧊 Temperature vs Ice Cream Sales\nr = {manual_corr:.3f} (Very Strong Positive)', fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Add data point labels
for i, (x, y) in enumerate(zip(temperature, ice_cream_sales)):
    ax1.annotate(f'Day {i+1}\n({x}°C, {y})', (x, y), xytext=(5, 5), 
                textcoords='offset points', fontsize=8, alpha=0.7)

# Plot 2: Detailed calculation table visualization
ax2 = axes[1]
ax2.axis('tight')
ax2.axis('off')

# Create table data
table_data = []
for i in range(len(temperature)):
    row = [
        f'Day {i+1}',
        f'{temperature[i]}',
        f'{ice_cream_sales[i]}',
        f'{temperature[i] - np.mean(temperature):.1f}',
        f'{ice_cream_sales[i] - np.mean(ice_cream_sales):.1f}',
        f'{(temperature[i] - np.mean(temperature)) * (ice_cream_sales[i] - np.mean(ice_cream_sales)):.1f}'
    ]
    table_data.append(row)

# Add summary row
table_data.append(['Sum', '', '', '0.0', '0.0', f'{np.sum((temperature - np.mean(temperature)) * (ice_cream_sales - np.mean(ice_cream_sales))):.1f}'])

column_labels = ['Day', 'Temp\n(°C)', 'Sales\n(units)', 'Temp\nDev', 'Sales\nDev', 'Product']

table = ax2.table(cellText=table_data, colLabels=column_labels, cellLoc='center', loc='center')
table.auto_set_font_size(False)
table.set_fontsize(9)
table.scale(1.2, 1.8)

# Style the table
for i in range(len(column_labels)):
    table[(0, i)].set_facecolor('#40466e')
    table[(0, i)].set_text_props(weight='bold', color='white')

# Highlight sum row
for i in range(len(column_labels)):
    table[(len(table_data), i)].set_facecolor('#f1f1f2')
    table[(len(table_data), i)].set_text_props(weight='bold')

ax2.set_title('📋 Detailed Calculations Table', fontweight='bold', pad=20)

plt.tight_layout()
plt.show()

# Summary statistics
print(f"\n📈 FINAL ANALYSIS SUMMARY")
print("=" * 50)
print(f"🌡️  Temperature Range: {np.min(temperature)}°C to {np.max(temperature)}°C")
print(f"🧊 Sales Range: {np.min(ice_cream_sales)} to {np.max(ice_cream_sales)} units")
print(f"📊 Sample Covariance: {manual_sample_cov:.2f}")
print(f"🎯 Correlation Coefficient: {manual_corr:.6f}")
print(f"💪 Relationship Strength: Very Strong Positive")
print(f"📋 Business Insight: Temperature explains {(manual_corr**2)*100:.1f}% of sales variation")
print(f"🔮 Prediction: For every 1°C increase, expect ~{np.polyfit(temperature, ice_cream_sales, 1)[0]:.1f} more sales")
````
