```python
import pandas as pd

# creating a simple Series in pandas

s = pd.Series([4, 7, -5, 3])

print(s)

print(type(s))

print("Index:", s.index)

print("Values:", s.values)

print("Data Type:", s.dtype)

print("Shape:", s.shape)

print("Number of Dimensions:", s.ndim)

print("Size:", s.size)
```

```python
import pandas as pd

s = pd.Series([4, 7, -5, 3], index=['d', 'b', 'a', 'c'])

print(s)

print(type(s))

print("Index:", s.index)

print("Values:", s.values)

print("Data Type:", s.dtype)

print("Shape:", s.shape)

print("Number of Dimensions:", s.ndim)

print("Size:", s.size)

print("Accessing element with index 'a':", s['a'])

print("Accessing element with index 2:", s[2])

print("Accessing multiple elements with indices 'c' and 'd':", s[['c', 'd']])
```

```python
import pandas as pd

arr = pd.Series(range(-10, 10, 2))

print(arr)

print(arr[arr > 0]) # Filter elements greater than 0

print(arr[arr <= 0]) # Filter elements less than or equal to 0

print(arr[arr % 3 == 0]) # Filter elements divisible by 3

print(arr * 2)
```

```python
import pandas as pd

sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}

s = pd.Series(sdata)

print(s)

print("Accessing element with index 'Texas':", s['Texas'])

print("Accessing multiple elements with indices 'Ohio' and 'Oregon':", s[['Ohio', 'Oregon']])

s['California'] = 2000

print("to dict:", s.to_dict())

states = ['California', 'Ohio', 'Oregon', 'Texas', 'Utah']

s2 = pd.Series(sdata, index=states)

print(s2)  # Note that 'California' will have NaN value

pd.isna(s2)  # Check for NaN values

pd.notna(s2)  # Check for non-NaN values
```
