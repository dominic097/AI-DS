```python
# Distance matrix between 5 cities
cities = ['A', 'B', 'C', 'D', 'E']
distance = {
    ('A','B'): 10, ('A','C'): 8, ('A','D'): 9, ('A','E'): 7,
    ('B','C'): 10, ('B','D'): 5, ('B','E'): 6,
    ('C','D'): 8, ('C','E'): 9,
    ('D','E'): 6
}

# Function to get distance (symmetric)
def get_distance(city1, city2):
    if city1 == city2: return 0
    if (city1, city2) in distance: return distance[(city1, city2)]
    return distance[(city2, city1)]

# Greedy TSP
def greedy_tsp(start):
    unvisited = set(cities)
    unvisited.remove(start)
    path = [start]
    total_cost = 0
    current = start

    while unvisited:
        next_city = min(unvisited, key=lambda c: get_distance(current, c))
        total_cost += get_distance(current, next_city)
        path.append(next_city)
        unvisited.remove(next_city)
        current = next_city

    # Return to start
    total_cost += get_distance(current, start)
    path.append(start)
    return path, total_cost

# Run Greedy TSP
path, cost = greedy_tsp('A')
print("Greedy TSP Path:", " -> ".join(path))
print("Total Cost:", cost)
```
