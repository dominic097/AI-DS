```python
import math

# Define the attack-defense game tree (as nested dictionary)
# Values represent the network safety score (higher = safer)
game_tree = {
    'root': ['Defense_A', 'Defense_B'],       # Defender’s choices
    'Defense_A': ['Attack_A1', 'Attack_A2'],  # Attacker’s responses
    'Defense_B': ['Attack_B1', 'Attack_B2'],

    # Leaf nodes — possible outcomes (safety score)
    'Attack_A1': [3, 5],
    'Attack_A2': [6, 9],
    'Attack_B1': [1, 2],
    'Attack_B2': [0, -1]
}

# Alpha-Beta Pruning implementation
def alpha_beta(node, depth, isMaximizing, alpha, beta):
    # Base condition: if leaf node (list of values)
    if isinstance(node, list):
        return max(node) if isMaximizing else min(node)
    
    # If node name points to a leaf (values)
    if node in game_tree and isinstance(game_tree[node][0], int):
        return max(game_tree[node]) if isMaximizing else min(game_tree[node])
    
    # If node name points to subnodes
    if node in game_tree:
        if isMaximizing:  # Defender’s turn
            best = -math.inf
            for child in game_tree[node]:
                val = alpha_beta(child, depth + 1, False, alpha, beta)
                best = max(best, val)
                alpha = max(alpha, best)
                if beta <= alpha:
                    print(f"Pruned branch at {child} (β ≤ α)")
                    break  # Beta Cutoff
            return best
        else:  # Attacker’s turn
            best = math.inf
            for child in game_tree[node]:
                val = alpha_beta(child, depth + 1, True, alpha, beta)
                best = min(best, val)
                beta = min(beta, best)
                if beta <= alpha:
                    print(f"Pruned branch at {child} (α ≥ β)")
                    break  # Alpha Cutoff
            return best
    return 0

# Run Alpha-Beta pruning from root
final_score = alpha_beta('root', 0, True, -math.inf, math.inf)
print("\nOptimal Network Safety Score:", final_score)
```
