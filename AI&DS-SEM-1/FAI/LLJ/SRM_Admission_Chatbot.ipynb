## Step 9: Analyze Chatbot Performance

Review conversation history and analyze common query patterns.

```python
# Analyze conversation history
history = chatbot.get_conversation_history()

print("📊 Chatbot Performance Analysis")
print("=" * 80)
print(f"\nTotal Interactions: {len(history)}")

if history:
    # Count intents
    intent_counts = {}
    for entry in history:
        intent = entry['intent']
        intent_counts[intent] = intent_counts.get(intent, 0) + 1
    
    print("\n📈 Intent Distribution:")
    print("-" * 80)
    for intent, count in sorted(intent_counts.items(), key=lambda x: x[1], reverse=True):
        percentage = (count / len(history)) * 100
        print(f"{intent.capitalize():<20s}: {count:>3d} queries ({percentage:>5.1f}%)")
    
    print("\n📝 Recent Conversations:")
    print("-" * 80)
    for i, entry in enumerate(history[-5:], 1):  # Show last 5 conversations
        print(f"\n[Conversation {i}]")
        print(f"Time: {entry['timestamp'].strftime('%Y-%m-%d %H:%M:%S')}")
        print(f"User: {entry['user']}")
        print(f"Intent: {entry['intent']}")
        print(f"Bot Response Preview: {entry['bot'][:100]}...")
else:
    print("\n⚠️ No conversation history available yet. Start chatting to generate data!")

print("\n" + "=" * 80)
```
