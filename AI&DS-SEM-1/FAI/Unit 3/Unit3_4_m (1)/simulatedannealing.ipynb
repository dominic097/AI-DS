```python
import math
import random

# -----------------------------
# Objective Function
# -----------------------------
# Function we want to minimize
def objective_function(x):
    return x**2 - 4*x + 4  # Quadratic function (Minimum at x=2)

# -----------------------------
# Simulated Annealing Algorithm
# -----------------------------
def simulated_annealing():
    # Step 1: Start with a random initial solution
    current_solution = random.uniform(-10, 10)
    current_value = objective_function(current_solution)

    # Step 2: Keep track of the best solution found so far
    best_solution = current_solution
    best_value = current_value    
    print(best_solution)
    print(current_value)

    # Step 3: Initialize parameters
    temperature = 100.0      # Initial temperature
    cooling_rate = 0.95      # How fast the temperature decreases

    # Step 4: Main loop - continue until temperature is very low
    while temperature > 1e-3:
        # Generate a new solution by small random change
        new_solution = current_solution + random.uniform(-1, 1)
        new_value = objective_function(new_solution)

        # Step 5: Decide whether to accept the new solution
        # Accept if it improves objective OR with some probability if worse
        if new_value < best_value or random.uniform(0, 1) < math.exp((current_value - new_value) / temperature):
            current_solution = new_solution
            current_value = new_value            

            # Update best solution if improvement
            if new_value < best_value:
                best_solution = new_solution
                best_value = new_value        

        # Step 6: Decrease temperature
        temperature *= cooling_rate    

    # Step 7: Return the best solution found
    return best_solution, best_value

# -----------------------------
# Run the algorithm
# -----------------------------
best_solution, best_value = simulated_annealing()

print(f"Best Solution: {best_solution}")
print(f"Best Value: {best_value}")
```
