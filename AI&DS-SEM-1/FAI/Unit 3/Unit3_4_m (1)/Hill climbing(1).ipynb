```python
import random

# Objective Function (to maximize)
def objective_function(x):
    return -(x-3)**2 + 9

# Hill Climbing Algorithm
def hill_climbing(max_iterations=1000, step_size=0.1):
    # Step 1: Start with a random solution
    current_solution = random.uniform(-10, 10)
    current_value = objective_function(current_solution)

    for _ in range(max_iterations):
        # Step 2: Generate a neighboring solution
        neighbor = current_solution + random.uniform(-step_size, step_size)
        neighbor_value = objective_function(neighbor)

        # Step 3: Move to neighbor if it's better
        if neighbor_value > current_value:
            current_solution = neighbor
            current_value = neighbor_value

    return current_solution, current_value

# -----------------------------
# Run Hill Climbing
# -----------------------------
best_solution, best_value = hill_climbing()
print(f"Best Solution: {best_solution}")
print(f"Best Value: {best_value}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 400)
y = objective_function(x)

plt.plot(x, y, label='Objective Function')
plt.plot(best_solution, best_value, 'ro', label='Hill Climbing Result')
plt.title("Hill Climbing Algorithm")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
