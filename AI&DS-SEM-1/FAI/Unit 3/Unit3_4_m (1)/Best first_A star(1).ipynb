```python
 import heapq

# Graph representation (Romania map simplified)
graph = {
    'Arad': {'Zerind': 75, 'Sibiu': 140, 'Timisoara': 118},
    'Zerind': {'Arad': 75, 'Oradea': 71},
    'Oradea': {'Zerind': 71, 'Sibiu': 151},
    'Sibiu': {'Arad': 140, 'Oradea': 151, 'Fagaras': 99, 'Rimnicu Vilcea': 80},
    'Timisoara': {'Arad': 118, 'Lugoj': 111},
    'Lugoj': {'Timisoara': 111, 'Mehadia': 70},
    'Mehadia': {'Lugoj': 70, 'Drobeta': 75},
    'Drobeta': {'Mehadia': 75, 'Craiova': 120},
    'Rimnicu Vilcea': {'Sibiu': 80, 'Craiova': 146, 'Pitesti': 97},
    'Fagaras': {'Sibiu': 99, 'Bucharest': 211},
    'Pitesti': {'Rimnicu Vilcea': 97, 'Craiova': 138, 'Bucharest': 101},
    'Craiova': {'Drobeta': 120, 'Rimnicu Vilcea': 146, 'Pitesti': 138},
    'Bucharest': {'Fagaras': 211, 'Pitesti': 101}
}

# Heuristic values (straight-line distance to Bucharest)
heuristic = {
    'Arad': 366, 'Bucharest': 0, 'Craiova': 160, 'Drobeta': 242,
    'Fagaras': 176, 'Lugoj': 244, 'Mehadia': 241, 'Oradea': 380,
    'Pitesti': 100, 'Rimnicu Vilcea': 193, 'Sibiu': 253,
    'Timisoara': 329, 'Zerind': 374
}

# -------------------------
# Best-First Search
# -------------------------
def best_first_search(start, goal):
    visited = set()
    queue = [(heuristic[start], start, [start])] #The queue stores tuples: (heuristic value, current city, path taken so far).
    while queue:
        (h, node, path) = heapq.heappop(queue) #pops the city with the smallest heuristic value h (closest to goal).
        if node == goal:
            return path
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                heapq.heappush(queue, (heuristic[neighbor], neighbor, path + [neighbor]))
    return None

# -------------------------
# A* Search
# -------------------------
def a_star_search(start, goal):
    open_list = [(heuristic[start], 0, start, [start])]
    visited = {}
    while open_list:
        f, g, node, path = heapq.heappop(open_list)
        if node == goal:
            return path, g
        if node in visited and visited[node] <= g:
            continue
        visited[node] = g
        for neighbor, cost in graph[node].items():
            new_g = g + cost
            new_f = new_g + heuristic[neighbor]
            heapq.heappush(open_list, (new_f, new_g, neighbor, path + [neighbor]))
    return None, float('inf')

# -------------------------
# Run and Compare
# -------------------------
start, goal = 'Arad', 'Bucharest'
print("Best First Search Path:", best_first_search(start, goal))
path, cost = a_star_search(start, goal)
print("A* Search Path:", path)
print("Total Path Cost:", cost)
```
