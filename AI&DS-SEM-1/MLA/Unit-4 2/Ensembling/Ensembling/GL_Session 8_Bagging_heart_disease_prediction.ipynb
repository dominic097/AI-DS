<h3>Train a model using decision tree and then using bagging</h3>

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

scores = cross_val_score(DecisionTreeClassifier(random_state=0), X, y, cv=5)
scores.mean()
```

**Use bagging now with decision tree**

estimator=DecisionTreeClassifier(random_state=0): Here, you're specifying the base estimator for the BaggingClassifier, which is a DecisionTreeClassifier. The random_state=0 ensures that the results are reproducible (i.e., the same random decisions will be made every time the code is run).

n_estimators=100: This parameter specifies the number of base models (decision trees) to train in the ensemble. Here, you're using 100 decision trees to create the bagging model. More trees generally improve the model's accuracy and robustness.

max_samples=0.9: This parameter controls the fraction of the training data used for each base estimator (each decision tree). Setting max_samples=0.9 means that each tree will be trained on 90% of the data. The remaining 10% of data will be left out and can be used for evaluating the model's performance (this is related to the out-of-bag score, explained below).

oob_score=True: This enables the out-of-bag (OOB) score, which is a built-in validation technique. When a decision tree is trained on a bootstrapped subset of the data (i.e., with replacement), some samples are left out (not included in the bootstrap sample). These samples are called "out-of-bag" samples. The OOB score is computed using these samples as a validation set, and it is used to estimate the generalization performance of the model without needing a separate validation set. This provides a way to assess model performance during training without needing to use cross-validation or a hold-out set.

random_state=0: This ensures that the random processes used in creating bootstrapped samples and random feature selections in each decision tree are reproducible. This is important for ensuring consistency in experiments.

```python
from sklearn.ensemble import BaggingClassifier
bag_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(random_state=0),
    n_estimators=100,
    max_samples=0.9,
    oob_score=True,
    random_state=0
)

scores = cross_val_score(bag_model, X, y, cv=5)
scores.mean()
```

**You can see that with bagging the score improved from 71.93% to 80.37%**

<h3>Train a model using Random Forest which itself uses bagging underneath</h3>

```python
from sklearn.ensemble import RandomForestClassifier

scores = cross_val_score(RandomForestClassifier(), X, y, cv=5)
scores.mean()
```

**Random forest gave even a better performance. Underneath it used bagging where it sampled not only data rows but also the columns (or features)**
